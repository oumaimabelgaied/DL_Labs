##imports

```python
import numpy as np
from keras.datasets import mnist
from keras.utils import to_categorical


from keras import backend as K

# fix random seed for reproducibility
seed = 7
np.random.seed(seed)

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

##Load Data

```python
def get_data_mnist():
    #load data
    (X_train, y_train), (X_test, y_test) = mnist.load_data()

    # reshape to be [samples][height][width][channels]
    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
    X_test  = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')

    #outputs
    num_classes = 10
    y_train = to_categorical(y_train, num_classes)
    y_test  = to_categorical(y_test, num_classes)

    return (X_train, y_train), (X_test, y_test), num_classes
```

##CNN model construction

```python
def small_model(num_classes):

    # create model
    model = Sequential()
    model.add(Conv2D(64, (3, 3), input_shape=(28, 28, 1), activation='relu'))
    model.add(Conv2D(32, (3, 3), activation='relu'))
    model.add(Flatten())
    model.add(Dense(num_classes, activation='softmax'))

    # compile model
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    return model
```

##Model training

```python
(X_train, y_train), (X_test, y_test), num_classes = get_data_mnist()

model = small_model(num_classes)

model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=200
)
```

##Evaluation of the model

```python
def print_model_error_rate(model, X_test, y_test):
  # Final evaluation of the model
  scores = model.evaluate(X_test, y_test, verbose=0)
  print("Model score : %.2f%%" % (scores[1]*100))
  print("Model error rate : %.2f%%" % (100-scores[1]*100))
```

```python
print_model_error_rate(model, X_test, y_test)
```

##Construction des 3 modГЁles aprГЁs normalisation de donnГ©es

```python
from keras.models import model_from_json
```

```python
# data WITH normalization
def normalized_data_mnist():
    #load data
    (X_train, y_train), (X_test, y_test) = mnist.load_data()

    #reshape data
    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
    X_test  = X_test.reshape(X_test.shape[0], 28, 28, 1)

    # Normalization
    X_train = X_train.astype('float32') / 255.0
    X_test  = X_test.astype('float32') / 255.0

    #one hot encode outputs
    num_classes = 10
    y_train = to_categorical(y_train, num_classes)
    y_test  = to_categorical(y_test, num_classes)

    return (X_train, y_train), (X_test, y_test), num_classes

# medium_CNN
def medium_CNN(num_classes):
    model = Sequential()
    model.add(Conv2D(32, (5, 5), input_shape=(28, 28, 1), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# large_CNN
def large_CNN(num_classes):
    model = Sequential()
    model.add(Conv2D(30, (5, 5), input_shape=(28, 28, 1), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(15, (3, 3), activation='relu'))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Save
def save_keras_model(model, filename):
    model_json = model.to_json()
    with open(filename + ".json", "w") as json_file:
        json_file.write(model_json)
    model.save_weights(filename + ".weights.h5")
    print(f"Model saved as {filename}.json and {filename}.weights.h5")

def load_keras_model(filename):
    # BE CAREFUL: model MUST be compiled before use!
    json_file = open(filename + ".json", 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    loaded_model = model_from_json(loaded_model_json)
    loaded_model.load_weights(filename + ".weights.h5")
    return loaded_model

# ------ Execution ------------------
(X_train2, y_train2), (X_test2, y_test2), num_classes2 = normalized_data_mnist()

# Execution small model with normalized data
print("=" * 40)
print("SMALL MODEL")
small = small_model(num_classes2)
small.fit(X_train2, y_train2, validation_data=(X_test2, y_test2), epochs=10, batch_size=200)
scores = small.evaluate(X_test2, y_test2, verbose=0)
print(f"Error rate: {100 - scores[1]*100:.2f}%")

# 3. medium_CNN
print("=" * 40)
print("MEDIUM CNN")
medium = medium_CNN(num_classes2)
medium.fit(X_train2, y_train2, validation_data=(X_test2, y_test2), epochs=10, batch_size=200)
scores = medium.evaluate(X_test2, y_test2, verbose=0)
print(f"Error rate: {100 - scores[1]*100:.2f}%")
save_keras_model(medium, "medium_cnn")

# 4. large_CNN
print("=" * 40)
print("LARGE CNN")
large = large_CNN(num_classes2)
large.fit(X_train2, y_train2, validation_data=(X_test2, y_test2), epochs=10, batch_size=200)
scores = large.evaluate(X_test2, y_test2, verbose=0)
print(f"Error rate: {100 - scores[1]*100:.2f}%")
save_keras_model(large, "large_cnn")

# Load example
loaded_large = load_keras_model("large_cnn")
loaded_large.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
scores = loaded_large.evaluate(X_test2, y_test2, verbose=0)
print(f"Loaded large CNN error rate: {100 - scores[1]*100:.2f}%")
```
